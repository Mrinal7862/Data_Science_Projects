```python
yes_no_columns = [
    "Partner",
    "Dependents",
    "PhoneService",
    "MultipleLines",
    "OnlineSecurity",
    "OnlineBackup",
    "DeviceProtection",
    "TechSupport",
    "StreamingTV",
    "StreamingMovies",
    "PaperlessBilling",
    "Churn"
]
```

```python
df1['yes_no_columns'] = df1['yes_no_columns'].replace({"Yes": 1, "No": 0})
df1['yes_no_columns'] = df1["gender"].replace({"Female": 1, "Male":0})
```

```python
df2 = pd.get_dummies(df1, columns=['InternetService', 'Contract', 'PaymentMethod'], dtype='int64')
df2.columns
```

```python
df2.sample(4)
```

```python
df2.dtypes
```

```python
df2.drop("yes_no_columns", axis=1, inplace=True)
```

```python
yes_no_columns = [
    "Partner",
    "Dependents",
    "PhoneService",
    "MultipleLines",
    "OnlineSecurity",
    "OnlineBackup",
    "DeviceProtection",
    "TechSupport",
    "StreamingTV",
    "StreamingMovies",
    "PaperlessBilling",
    "Churn"
]

df2[yes_no_columns] = df2[yes_no_columns].replace({
    "Yes": 1,
    "No": 0
})
```

```python
df2[yes_no_columns] = df2[yes_no_columns].astype("int64")
```

```python
df2["gender"] = df2["gender"].map({
    "Male": 1,
    "Female": 0
})
```

```python
df2.dtypes
```

```python
cols_to_scale = ['tenure', 'MonthlyCharges', 'TotalCharges']

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
```

```python
df2[cols_to_scale] = scaler.fit_transform(df2[cols_to_scale])
```

```python
df2
```

```python
df2.sample(3)
```

```python
X = df2.drop('Churn', axis='columns')
y = df2["Churn"] 
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
import tensorflow as tf 
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Dense(20, input_shape= (26,), activation='relu'),
    
    keras.layers.Dense(1,activation='sigmoid'),
    
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=100)
```

```python
model.evaluate(X_test, y_test)
```

```python
yp = model.predict(X_test)
```

```python
yp[:5]
```

```python
y_pred = []
for element in yp:
    if element > 0.5: 
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
y_pred[:10]
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(classification_report(y_test, y_pred))
```

```python
import seaborn as sns 
cm = tf.math.confusion_matrix(labels=y_test, predictions=y_pred)

plt.figure(figsize = (10, 7))
sns.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted'  )
plt.ylabel('Truth')
```
