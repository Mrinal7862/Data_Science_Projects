## Time-based features

### Subtask:
Extract features from the time columns, such as hour of the day, day of the week, or time duration between events.


```python
dfc['creation_time'] = pd.to_datetime(dfc['creation_time'])
dfc['end_time'] = pd.to_datetime(dfc['end_time'])
dfc['creation_hour'] = dfc['creation_time'].dt.hour
dfc['creation_day_of_week'] = dfc['creation_time'].dt.dayofweek
dfc['duration'] = (dfc['end_time'] - dfc['creation_time']).dt.total_seconds()
display(dfc.head())
```

## Interaction features




**Reasoning**:
Create interaction terms between existing numerical features by multiplying bytes_in with bytes_out and creation_time with end_time, then display the head of the dataframe.



```python
dfc['bytes_interaction'] = dfc['bytes_in'] * dfc['bytes_out']
dfc['time_interaction'] = dfc['creation_time'] * dfc['end_time']
display(dfc.head())
```

**Reasoning**:
The multiplication of datetime objects failed. Convert the time columns to numerical representation (Unix timestamp) before performing multiplication.



```python
dfc['creation_time'] = dfc['creation_time'].astype('int64') // 10**9
dfc['end_time'] = dfc['end_time'].astype('int64') // 10**9
dfc['bytes_interaction'] = dfc['bytes_in'] * dfc['bytes_out']
dfc['time_interaction'] = dfc['creation_time'] * dfc['end_time']
display(dfc.head())
```

## Polynomial features

### Subtask:
Create polynomial features from numerical columns to capture non-linear relationships.


**Reasoning**:
Identify and select numerical columns from the DataFrame to create polynomial features.



```python
numerical_cols = dfc.select_dtypes(include=np.number).columns.tolist()
print("Numerical columns:", numerical_cols)
selected_cols = ['bytes_in', 'bytes_out', 'duration']
```

**Reasoning**:
Create polynomial features for the selected numerical columns and add them to the DataFrame.



```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2, include_bias=False, interaction_only=False)
poly_features = poly.fit_transform(dfc[selected_cols])
poly_feature_names = poly.get_feature_names_out(selected_cols)

poly_df = pd.DataFrame(poly_features, columns=poly_feature_names)
dfc = pd.concat([dfc, poly_df], axis=1)
display(dfc.head())
```

**Reasoning**:
Create new ratio features by dividing bytes_out by bytes_in and vice versa, handling potential division by zero.



# Task
Apply anomaly detection to the dataframe `dfc`.

## Select features



```python
features = [
    'bytes_in', 'bytes_out', 'creation_time', 'end_time', 'src_ip',
    'response.code', 'dst_port', 'dst_ip', 'rule_names', 'time',
    'src_ip_country_code_AE', 'src_ip_country_code_AT',
    'src_ip_country_code_CA', 'src_ip_country_code_DE',
    'src_ip_country_code_IL', 'src_ip_country_code_NL',
    'src_ip_country_code_US', 'protocol_HTTPS', 'creation_hour',
    'creation_day_of_week', 'duration', 'bytes_interaction',
    'time_interaction', 'bytes_in^2', 'bytes_in bytes_out',
    'bytes_in duration', 'bytes_out^2', 'bytes_out duration',
    'duration^2', 'bytes_ratio', 'in_out_ratio'
]

df_selected = dfc[features]
display(df_selected.head())
```

## Scale data




**Reasoning**:
Import StandardScaler from sklearn.preprocessing and instantiate it. Then, fit and transform the selected features and store them in a new DataFrame with the original column names.



```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df_selected), columns=df_selected.columns)
display(df_scaled.head())
```

## Choosing and train an anomaly detection model

**Reasoning**:
Import Isolation Forest, instantiate the model, and train it on the scaled data.



```python
from sklearn.ensemble import IsolationForest

model = IsolationForest(random_state=42)
model.fit(df_scaled)
```

## Predict anomalies

### Subtask:
Use the trained model to predict anomaly scores or labels for the data.

```python
print("Distribution of Anomaly Labels:")
display(dfc['anomaly_label'].value_counts())

print("\nCharacteristics of Anomalies (anomaly_label == -1):")
display(dfc[dfc['anomaly_label'] == -1].describe())

print("\nCharacteristics of Inliers (anomaly_label == 1):")
display(dfc[dfc['anomaly_label'] == 1].describe())
```

**Reasoning**:
Predict anomaly labels and scores, add them to the original dataframe, and display the head.



```python
dfc['anomaly_label'] = model.predict(df_scaled)
dfc['anomaly_score'] = model.decision_function(df_scaled)
display(dfc.head())
```

## Evaluate and interpret results




```python
import matplotlib.pyplot as plt
import seaborn as sns

# Remove duplicate columns from df_selected before plotting
dfc_unique_cols = dfc.loc[:,~dfc.columns.duplicated()]


plt.figure(figsize=(12, 6))
sns.scatterplot(x='bytes_in', y='anomaly_score', hue='anomaly_label', data=dfc_unique_cols)
plt.title('Anomaly Score vs Bytes In')
plt.xlabel('Bytes In')
plt.ylabel('Anomaly Score')
plt.show()

plt.figure(figsize=(12, 6))
sns.scatterplot(x='bytes_out', y='anomaly_score', hue='anomaly_label', data=dfc_unique_cols)
plt.title('Anomaly Score vs Bytes Out')
plt.xlabel('Bytes Out')
plt.ylabel('Anomaly Score')
plt.show()
```
