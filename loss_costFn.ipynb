```python
import numpy as np

y_predicted = np.array([1,1,0,0,1])
y_actual = np.array([0.30, 0.7,1, 0, 0.5])
```

```python
def mae(y_act, ypred):
  total_err = 0
  for yt, yp in zip(y_act, ypred):
    total_err += abs(yt - yp)
  print("Total error: ", total_err)

  mae = total_err / len(y_act)
  print("MAE: ", mae)
  return mae
```

```python
mae(y_actual, y_predicted)
```

```python
np.mean(np.abs(y_predicted - y_actual))
```

```python
# MSE  >> using brute force

def mse(yi, yp):
  lst = 0
  for ya, ypr in zip(yi, yp):
    lst += (ypr - ya)**2
  print("squared: " , lst)

  mse = lst / len(yi)
  print("MSE: ", mse)
```

```python
mse([8], [9])
```

```python
# MSE using numpy
import numpy as np

def msenp(yi, yp):
  mse = np.mean(np.square(yi - yp))
  print("MSE: " , mse)
```

```python
msenp(8, 9)
```
